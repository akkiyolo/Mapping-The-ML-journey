```python
### Merging and joining Dataframes
# Create sample DataFrames
df1= pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1, 2, 3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value2': [4, 5, 6]})
```

```python
df1
```

```python
df2
```

```python
## merge data frames on the 'key colummns

pd.merge(df1,df2,on="Key",how="inner")

## Matches rows where the value in the "Key" column is identical in both df1 and df2.
```

```python
## All distinct keys that appear in either df1 or df2 are kept.

pd.merge(df1,df2,on="Key",how="outer")
```

```python
## more priority to the left join will be given 
## Keep every row from df1 (the left DataFrame), regardless of whether its "Key" exists in df2.
pd.merge(df1,df2,on="Key",how="left")
```

```python
## more priority to the right join will be given
## Keep every row from df2 (the right DataFrame), regardless of whether its "Key" exists in df1.
pd.merge(df1,df2,on="Key",how="right")
```
